```python
###factorial function 
def factorial(num):
    d=[]
    return 1 if num==1 else (num*factorial(num-1))
num=5
print("Factorial of {0} is {1}".format(num,factorial(num)))
```

###simple double function

```python
def double(x):
    return x*2

print(double(5))
```

```python
###lambda function
double = lambda x : x*2
print(double(5))
```

```python
###example use with filter()
lst=[1,2,3,4,5]
even_list=list(filter(lambda x:(x%2==0),lst))
print(even_list)
```

```python
###normal function
def double(x): return x *2

print(double(5))
```

```python
###lambda function
double=lambda x : x*2
print(double(5))
```

```python
from functools import reduce

lst=[1,2,3,4,5]
product_lst=reduce(lambda x,y : x*y,lst)
print(product_lst)
```
