```python
# Sine Wave plot
import numpy as np
from matplotlib import pyplot as plt

###Compute the x and y coordiantes for points on a sine wave
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.title('Sine Wave Form')
### Plots the point using matplotlib
plt.plot(x,y)
plt.show()
```

```python
####Subplot function allows to plot different things in the same figure

import numpy as np
from matplotlib import pyplot as plt

##Compute the x and y cooridnates for points on sine and cosine curves
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

##Setup  a subplot grid that has height 2 and widht 1
## and set the first such plot as active

plt.subplot(2,1,1)

##Make the first plot
plt.plot(x,y_sin)
plt.title('Sine')

###Set the second subplot as active and mmake the second plot 

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

#Show the figure

plt.show()
```

Bar()
pyplot provides bar() function to generate bar graphs 

```python
from matplotlib import pyplot as plt

x=[5,8,12]
y=[12,16,6]

x2=[6,9,11]
y2=[6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g', align='center')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

numpy.histogram() function that is a graphical representation of the frequency distribution of data
It has two parameters which is input array and bins parameters

```python
import numpy as np

a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
np.histogram(a,bins=[0,20,40,60,80,100])
hist,bins=np.histogram(a,bins=[0,20,40,60,80,100])
print(hist)
print(bins)
```

Matplotlib can convert this numeric representation of histogram into a graph and takes the array containing the data  and 
bin array as parameters and converts into a histogram

```python
from matplotlib import pyplot as plt
import numpy as np

a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=[0,20,40,60,80,100])
plt.title('Histogram')
plt.show()
```
