# Matplotlib in python

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.xlabel('X label')
plt.show()
```

```python
import numpy as np

##evenly sampled time at 200ms intervals

t=np.arange(0.,5.,0.2)

##red dashes , blue squares and green triangles

plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

# Plotting with keywords string
matplotlib allows you provide such an object with the data keyword argument

```python
data={'a':np.arange(50),
      'c':np.random.randint(0,50,50),
      'd':np.random.randn(50)}
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100

plt.scatter('a','b',c='c',s='d',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

Plotting with categorical variables

It is also possible to create a plot using categorical variables , and allows to pass categorical variables directly to many plotting functions

```python
names=['group_a','group_b','group_c']
values=[1,10,100]
plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
plt.suptitle('Categorical Plotting')
plt.show()
```

###working with multiple figures and axes

Matlab, and pyplot have the concept of the current figure and the current axes

```python
def f(t):
    return np.exp(-t)* np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.2)

plt.figure()
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

Figure function can use multiple figures and can contain as many axes and subplots as ur desires
